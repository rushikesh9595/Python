## Removing outliers for sales variable and making it fairly skewed using log transformation

```python
sns.boxplot(x = df['sales'])

Q1_sales = df['sales'].quantile(0.25)
Q1_sales

Q2_sales = df['sales'].quantile(0.5)
Q2_sales

Q3_sales = df['sales'].quantile(0.75)
Q3_sales

IQR_sales = Q3_sales - Q1_sales
IQR_sales

upper_sales = Q3_sales+(1.5*IQR_sales)
upper_sales

lower_sales = Q1_sales-(1.5*IQR_sales)
lower_sales
```

```python
df['sales'] = np.where(df['sales']>498.93, min(df['sales']), df['sales'])
df

df['sales'].skew()         ## highly rightward skewed data (positive)

sns.boxplot(x = df['sales'])        ## Boxplot with removed outliers for 'sales' variable

sns.displot(df['sales'])

df['sales'] = np.where(df['sales']<-271.71, -271.71, df['sales'])
df

log_sales = np.log(df['sales'])        ## using log transformation to handle right skewed data
log_sales

log_sales.skew()          ## Fairly symmetrical data

sns.displot(log_sales)
```
