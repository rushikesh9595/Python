```python
# ### user defined functions

# Syntax
# def any_function_name(agru1,argu2...):    (Defining the function)
#     expression and statements
## function_name(var1,var2..)      (Calling the function)
```

```python
def sum(a,b):   #(Defining the function)
    c = a+b
    print(c)     ## print won't allow to use the stored c for future purposes

sum(30,30)     #(Calling the above function)
```

```python
#print won't allow to use the stored c for future purposes

k = sum(30,30)
print(k)     ## for print(k) its showing none
```

```python
#but if we use return c instead of print(c) it will help us use that c value for future
def sum(a,b):   #(Defining the function)
    c = a+b
    return c

k = sum(30,30)      ##### now its showing me the value because return c allows users to use the c outside the function but print doesnt
print(k)
```
