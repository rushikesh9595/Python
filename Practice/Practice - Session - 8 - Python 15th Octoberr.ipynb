```python
sns.boxplot(x = tip['total_bill'])    ### Total bill has an outlier
```

```python
Q1 = tip['total_bill'].quantile(0.25)
Q1
```

```python
Q2 = tip['total_bill'].quantile(0.5)
Q2
```

```python
Q3 = tip['total_bill'].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
Upper_Thre_Value = Q3 + (1.5*IQR)    ### It gives from where outlier starts
Upper_Thre_Value         #### for this outlier values starts from 40.29 
```

```python
tip[tip['total_bill']>40.29].sort_values('total_bill', ascending = True)
```

```python
#### if__else function using np.where

tip['total_bill'] = np.where(tip['total_bill']>40.29, 40.29, tip['total_bill'])
pd.DataFrame(tip['total_bill'])

### it is read as where total bill is greater than 40.29 mention it as 40.29 else if not return total_bill

    
```
